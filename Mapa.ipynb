```python
import folium
import matplotlib
class Mapa:
    def __init__(self, lat, lon, zoom):
        self.mapa = folium.Map(location=[lat, lon], zoom_start=zoom)
        self.camadas = []
        self.marcadores = []
    
    def adicionar_camada(self, camada):
        self.camadas.append(camada)
        camada.add_to(self.mapa)
    
    def adicionar_marcador(self, lat, lon, popup):
        marcador = folium.Marker(location=[lat, lon], popup=popup)
        self.marcadores.append(marcador)
        marcador.add_to(self.mapa)
    
    def exibir(self):
        folium.LayerControl().add_to(self.mapa)
        return self.mapa

#criar um objeto de mapa com latitude, longitude e zoom iniciais
m = Mapa(lat=-23.5489, lon=-46.6388, zoom=12)

#criar uma camada de marcadores para restaurantes
camada_restaurantes = folium.FeatureGroup(name='Restaurantes')
marcadores_restaurantes = [
(-23.5470, -46.6350, 'Restaurante A'),
(-23.5500, -46.6420, 'Restaurante B'),
(-23.5530, -46.6400, 'Restaurante C'),
]
for lat, lon, popup in marcadores_restaurantes:
    camada_restaurantes.add_child(folium.Marker(location=[lat, lon], popup=popup))

#criar uma camada de marcadores para museus
camada_museus = folium.FeatureGroup(name='Museus')
marcadores_museus = [
(-23.5520, -46.6360, 'Museu A'),
(-23.5480, -46.6390, 'Museu B'),
(-23.5500, -46.6440, 'Museu C'),
]
for lat, lon, popup in marcadores_museus:
    camada_museus.add_child(folium.Marker(location=[lat, lon], popup=popup))

#adicionar as camadas de marcadores ao mapa
m.adicionar_camada(camada_restaurantes)
m.adicionar_camada(camada_museus)

#exibir o mapa
m.exibir()

```
